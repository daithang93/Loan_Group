# Nhập dữ liệu PLN theo 780

```python
import os
os.chdir(r"\\10.64.196.31\3. Modelling\2. DU LIEU CA NHAN\Nho.pv\12. Validation project\7. Hieu chinh\KHDN\Corp2019\Data ITC\LN\Restructured under 780")
```

```python
import glob
all_file_780=[i for i in glob.glob("780*.xlsx")]
all_file_780[0:5]
```

```python
import pandas as pd 
ccList=[]
for f in all_file_780:
    cc=pd.read_excel(f)
    cc.columns = ['CIFNO', 'LN_GROUP', 'BUSS_DATE']
    ccList.append(cc)
ccDF = pd.concat(ccList, axis=0)
print(ccDF.shape)
print(ccDF.head())
```

```python
cc=ccDF
print(cc.dtypes)
print("-"*50)
print(cc.isnull().sum()) 
```

```python
cc['CIFNO']=cc.CIFNO.apply(str)
print(cc['CIFNO'].apply(type).value_counts())
cc
```

```python
cc[cc.BUSS_DATE.isnull()]
```

```python
cc1=cc.dropna(how="any")
print(cc1.isnull().sum())
```

```python
cc2=cc1[~(cc1.CIFNO=='Grand Total')]
cc2
```

```python
#reset index
cc3=cc2.reset_index(drop=True)
cc3
```

```python
# Do bảng LN780 có CIF và BUSS_DATE bị trùng nhưng nhóm nợ khác nhau => Lọc trùng lấy nhóm nợ cao nhất
cc3.sort_values(by="LN_GROUP", ascending=False, inplace=True)
cc4=cc3.drop_duplicates(subset=['CIFNO','BUSS_DATE'], keep='first').reset_index(drop=True)
cc4
```

```python
cc5=cc4.groupby(['CIFNO','BUSS_DATE']).agg({'LN_GROUP':"max"}).sort_values(by="LN_GROUP",ascending=False)
cc5
```

```python
df3
```

# Nhập dữ liệu PLN theo 780

```python
#Left join theo key CIFNO và BUSS_DATE
ln_all=df3.merge(cc5, how='left', left_index=True, right_index=True)
ln_all
```
